```python
# Gerando um DataFrame a partir da API do BCB
import sqlite3
import pandas as pd

conexao = sqlite3.connect('src/datasets/etlbcb.db')
consulta = 'select * from meios_pagamentos_tri'
dados = pd.read_sql(consulta, conexao)
conexao.close()

dados
```

```python
# Exibindo o resumo estatístico da única coluna qualitativa (categorica)
dados['datatrimestre'].describe()
```

```python
# Criando um novo DataFrame sem a coluna categorica (datatrimestre) do DataFrame
dados_numericos = dados.select_dtypes(exclude='object')

# Normalizar os valores para a mesma unidade de medida.
# Os valores estão e milhões e as quantidades estão em milhares.
# Normalizando os valores para milhares.
dados_numericos['valorPix'] = dados_numericos['valorPix'] * 1_000
dados_numericos['valorTED'] = dados_numericos['valorTED'] * 1_000
dados_numericos['valorTEC'] = dados_numericos['valorTEC'] * 1_000
dados_numericos['valorCheque'] = dados_numericos['valorCheque'] * 1_000
dados_numericos['valorBoleto'] = dados_numericos['valorBoleto'] * 1_000
dados_numericos['valorDOC'] = dados_numericos['valorDOC'] * 1_000
dados_numericos['valorCartaoCredito'] = dados_numericos['valorCartaoCredito'] * 1_000
dados_numericos['valorCartaoDebito'] = dados_numericos['valorCartaoDebito'] * 1_000
dados_numericos['valorCartaoPrePago'] = dados_numericos['valorCartaoPrePago'] * 1_000
dados_numericos['valorTransIntrabancaria'] = dados_numericos['valorTransIntrabancaria'] * 1_000
dados_numericos['valorConvenios'] = dados_numericos['valorConvenios'] * 1_000
dados_numericos['valorDebitoDireto'] = dados_numericos['valorDebitoDireto'] * 1_000
dados_numericos['valorSaques'] = dados_numericos['valorSaques'] * 1_000

dados_numericos
```

```python
# Média aritmética das colunas
dados_numericos.mean()
```

```python
# Média ponderada
media_ponderada = (dados_numericos * dados_numericos).sum() / dados_numericos.sum()
media_ponderada
```

```python
# Média harmônica
media_harmonica = len(dados_numericos) / (1 / dados_numericos).sum()
media_harmonica
```

```python
# Média geometrica
import numpy as np

media_geom = np.exp(np.log(dados_numericos).mean())
media_geom
```
